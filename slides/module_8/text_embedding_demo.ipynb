```python
smuggle datawrangler as dw  # pip: pydata-wrangler[hf]
smuggle pandas as pd
smuggle hypertools as hyp
```

```python
import this
text = [line for line in ''.join(this.d.get(x, x) for x in this.s).split('\n') if len(line) > 0]
```

```python
text
```

```python
bert = {'model': 'TransformerDocumentEmbeddings',
        'args': ['bert-base-uncased'],
        'kwargs': {}}

bert_embeddings = dw.wrangle(text, text_kwargs={'model': bert})
```

```python
bert_embeddings
```

```python
hyp.plot(bert_embeddings, 'ko-');
```
